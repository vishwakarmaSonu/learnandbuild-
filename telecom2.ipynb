```python
def bis(y):
    
    if y == 'Yes':
        return 1
    else:
        return 0
```

```python
data['Partner']= data['Partner'].apply(bis)
data['Dependents'] = data['Dependents'].apply(bis)
data['PhoneService'] = data['PhoneService'].apply(bis)
data['MultipleLines']  = data['MultipleLines'].replace('No phone service','No')
data['MultipleLines']=data['MultipleLines'].apply(bis)
data['OnlineSecurity'] = data['OnlineSecurity'].replace('No phone service','No')
data['OnlineSecurity'] =data['OnlineSecurity'].apply(bis)
data['OnlineBackup'] = data['OnlineBackup'].replace('No phone service','No')
data['OnlineBackup'] =data['OnlineBackup'].apply(bis)
data['DeviceProtection'] = data['DeviceProtection'].replace('No phone service','No')
data['DeviceProtection'] =data['DeviceProtection'].apply(bis)
data['TechSupport'] = data['TechSupport'].replace('No phone service','No')
data['TechSupport'] =data['TechSupport'].apply(bis)
data['StreamingTV'] = data['StreamingTV'].replace('No phone service','No')
data['StreamingTV'] =data['StreamingTV'].apply(bis)
data['StreamingMovies'] = data['StreamingMovies'].replace('No phone service','No')
data['StreamingMovies'] =data['StreamingMovies'].apply(bis)
data['PaperlessBilling'] =data['PaperlessBilling'].apply(bis)
data['Churn'] =data['Churn'].apply(bis)


```

```python
inter = {'DSL':0,'Fiber optic':1,'No':2}
data['InternetService']= data['InternetService'].map(inter)
```

```python
cont = {'Month-to-month':0,'One year':1,'Two year':2}
data['Contract']= data['Contract'].map(cont)
print(cont.keys())
# ['Month-to-month' 'One year' 'Two year']
```

```python
def vis(y):
    if y == 'Male':
        return 0
    else:
        return 1

data['gender'] = data['gender'].map(vis)
```
